```python
import pandas as pd
import sklearn as sklearn
import scipy.stats as stats
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import RandomizedSearchCV

from utils.randomized_search import pNpUniform
from utils.base_set import X_train, y_train, seed
```

### LDA 

```python
model = LinearDiscriminantAnalysis()
k = 5
cv = StratifiedKFold(n_splits=k, shuffle=True, random_state=seed)
params = {
    'solver': ["svd", "lsqr"], 
    'priors': pNpUniform(0, 1),
    'tol': stats.uniform(0, 1e-2),
    'store_covariance': [True,False]
}
n = 100
classifier_LDA = RandomizedSearchCV(estimator=model,
                                param_distributions=params,
                                cv=cv,
                                n_iter=n,
                                scoring="roc_auc",
                                random_state=seed)
classifier_LDA.fit(X_train, y_train)
```

```python
best = classifier_LDA.best_params_
auc_roc = classifier_LDA.best_score_
print(best)
print(auc_roc)
```

```python
columns_to_keep = ['param_solver', 'param_priors', 'param_tol','mean_test_score','rank_test_score']
pd.DataFrame(classifier_LDA.cv_results_).sort_values('rank_test_score')[columns_to_keep]
```
