```python
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold

from utils.metrics import aucroc
from utils.curves import learning_curve
from utils.estimators import SVCWrapper
from utils.base_set import X_train, y_train, seed
```

### SVM

```python
best_params = {'C': 2.5491739956475503, 'degree': 8, 'gamma': 'scale', 'kernel': 'rbf'} # based on ej. 3 y 4
k = 5
n = X_train.shape[0]
cv = StratifiedKFold(n_splits=k, shuffle=True, random_state=seed)
fig, ax = plt.subplots(figsize=(10, 6))
train = range(50, n, 10)
mean_val_scores, _, _, _ = learning_curve(SVCWrapper(**best_params, random_state=seed),
                 train,
                 X_train,
                 y_train,
                 aucroc,
                 True,
                 cv,
                 ax)
ax.legend(fontsize=20, loc='lower right')
ax.set_xlabel('Cantidad de datos', fontsize=20)
ax.set_ylabel('aucroc', fontsize=20)
ax.tick_params(axis="x", labelsize=16)
ax.tick_params(axis="y", labelsize=16)
ax.grid(True)
```
